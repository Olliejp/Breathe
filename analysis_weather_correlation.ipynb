```python
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

## Spatial Manipulation
import geopandas as gpd
from shapely.geometry import Point

## Mapping
import folium
from folium import plugins

## Misc
import functools
from itertools import product
import contextily as ctx

pd.options.display.max_columns = None
#pd.options.display.max_rows = None

from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
df = pd.read_csv('pm_no_clean.csv', index_col=0)
```

```python
pm25_cor = df[['pm2_5_ugm3', 'temperature', 'windBearing', 'windSpeed', 'pressure']].dropna()
```

```python
## Plot correlation matrix for pm25
cmap = sns.diverging_palette(220, 20, sep=20, as_cmap=True)
sns.clustermap(pm25_cor.corr(), figsize= (20,20), cmap = cmap, annot=True).ax_row_dendrogram.set_visible(False)
plt.show()
```
